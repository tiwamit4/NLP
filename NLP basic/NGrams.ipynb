```python
import pandas as pd
df=pd.read_csv('spam.csv', encoding='latin-1')
```

```python
df
```

```python
df = df.drop('Unnamed: 2', axis=1)
df = df.drop('Unnamed: 3', axis=1)
df = df.drop('Unnamed: 4', axis=1)
```

```python
df.rename(columns = {'v1':'label'}, inplace = True)
df.rename(columns = {'v2':'message'}, inplace = True)
```

```python
messages=df
messages
```

### Data Cleaning and preprocessing

```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

### create Bag Of Word Model

```python
from sklearn.feature_extraction.text import CountVectorizer
```

### UniGrams

```python
cv = CountVectorizer(max_features=100,binary=True)
x=cv.fit_transform(corpus).toarray()
x
```

```python
cv.vocabulary_
```

### uniGrams

```python
cv = CountVectorizer(max_features=100,binary=True,ngram_range=(1,1))
x=cv.fit_transform(corpus).toarray()
x
```

```python
cv.vocabulary_
```

### UniGrams + biGrams

```python
cv = CountVectorizer(max_features=1000,binary=True,ngram_range=(1,2))
x=cv.fit_transform(corpus).toarray()
x
```

```python
cv.vocabulary_
```

### UniGrams + biGrams + triGrams

```python
cv = CountVectorizer(max_features=1000,binary=True,ngram_range=(1,3))
x=cv.fit_transform(corpus).toarray()
x
```

```python
cv.vocabulary_
```
